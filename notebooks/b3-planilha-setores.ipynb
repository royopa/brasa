```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../data/Setorial B3 13-11-2023 (português).xlsx")
```

```python
rows = []
for ix in df.index:
    if df.iloc[ix, 0] == "SETOR ECONÔMICO":
        continue
    if not pd.isna(df.iloc[ix, 0]):
        sector = df.iloc[ix, 0]
    if not pd.isna(df.iloc[ix, 1]):
        subsector = df.iloc[ix, 1]
    if not pd.isna(df.iloc[ix, 2]) and pd.isna(df.iloc[ix, 3]) and pd.isna(df.iloc[ix, 4]):
        segment = df.iloc[ix, 2]
    if not pd.isna(df.iloc[ix, 2]) and not pd.isna(df.iloc[ix, 3]):
        row = (sector, subsector, segment, df.iloc[ix, 3])
        rows.append(row)

df_sectors = pd.DataFrame(rows, columns=["sector", "subsector", "segment", "asset_name"])
df_sectors
```

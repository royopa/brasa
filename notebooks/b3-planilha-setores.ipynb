```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../data/Setorial B3 28-03-2023 (português).xlsx")
```

```python
rows = []
for ix in df.index:
    if not pd.isna(df.iloc[ix, 0]) and df.iloc[ix, 0] != "SETOR ECONÔMICO":
        sector = df.iloc[ix, 0]
        subsector = df.iloc[ix, 1]
        segment = df.iloc[ix, 2]
    elif pd.isna(df.iloc[ix, 0]) and not pd.isna(df.iloc[ix, 2]):
        row = (sector, subsector, segment, df.iloc[ix, 3])
        rows.append(row)

pd.DataFrame(rows, columns=["sector", "subsector", "segment", "asset"])
```
